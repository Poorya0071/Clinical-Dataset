# Heart Disease Prediction

## Project Overview
This project aims to predict whether a person will have heart disease in the next ten years. It specifically tackles the overlooked issue of data imbalance found in previous studies of similar datasets, which often reported misleadingly high accuracy scores. This project corrects data imbalance and leakage, providing a more reliable predictive model.


## Data Source
The dataset is sourced from Kaggle and was previously used without addressing its imbalance issues. This project addresses these issues to ensure more valid results.

- [Heart Disease Prediction Using Logistic Regression - Kaggle Code](https://www.kaggle.com/code/neisha/heart-disease-prediction-using-logistic-regression)
- [Heart Disease Prediction Using Logistic Regression - Kaggle Dataset](https://www.kaggle.com/datasets/dileep070/heart-disease-prediction-using-logistic-regression/code)

## Handling Outliers


```python
sns.distplot(data_no_mv['cigsPerDay'])
# Analyze the distribution of cigarettes smoked per day
```

```python
q_cig = data_no_mv['cigsPerDay'].quantile(0.99)
q_cig
# Removing extreme outliers in cigarette consumption
```

```python
data_cig = data_no_mv[data_no_mv['cigsPerDay']<q_cig]
```

```python
sns.distplot(data_cig['cigsPerDay'])
```

```python
data_cig.describe()
# we droped data sets from 3987 to 3913 data points
```

```python
# Removing outliers based on medical knowledge for cholesterol and blood pressure

data_chol = data_cig[data_cig['totChol']<400]
```

```python
data_chol.describe()
# we removed 9 data points.
```

```python
# for sysBP, the maximum could be 220 based on reasearch.
data_sysBP = data_chol[data_chol['sysBP']<220]
```

```python
data_sysBP.describe()
```

## let's remove the outlires for the DiaBP, Heart rate, and BMI, according to the data on the research papers


```python
# diaBP outliers:
data_clean_1 = data_sysBP[data_sysBP['diaBP']<125]
```

```python
# BMI
data_clean_2 = data_clean_1[data_clean_1['BMI']<40]
```

```python
#HeartRate
data_clean_3 = data_clean_2[data_clean_2['heartRate']<130]
```

```python
data_clean_3.describe()
# in the process of removing missing data and outliers, I droped 380 data points, around 8% of total data points.
```

## # Prepare the data for modeling


```python
data_preprocessed = data_clean_3.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Define inputs and outputs
x = data_preprocessed.drop("TenYearCHD",axis=1)
y = data_preprocessed["TenYearCHD"]
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Display sizes of the splits
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier # import RandomForestClassifier as the model.
model = RandomForestClassifier()
model.fit(x_train , y_train)
```

## # Evaluate the model using accuracy and confusion matrix


```python
from sklearn.metrics import accuracy_score, confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
model.score(x_test, y_test)
```

```python
# let's predict the model performance on the test data set.
y_pred = model.predict(x_test)
```

```python
# let's find confusion_matrix
ConfusionMatrix = confusion_matrix(y_test, y_pred)
ConfusionMatrix
```

## **Leakage Correction**:
there is a leak in our data because the labels are imbalance. Let's deal with the data leakage.

```python
y.value_counts()
# the number of zero values are 6 times more than the number of one values. Therefore, we have data leakage.
# let's use RandomOverSampler to deal with the data implance problem. 
```

```python
# Identify and address data leakage due to label imbalance

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(
    sampling_strategy='auto', # samples only the minority class
    random_state=0,  # for reproducibility
)  

x_ros, y_ros = ros.fit_resample(x, y)
```

```python
y_ros.value_counts()
# wonderfull, we do not have any data leakage and data imbalanced now.
```

```python
# Re-train the model with corrected data
np.random.seed(42)
x_train,x_test, y_train, y_test = train_test_split(x_ros,y_ros,test_size=0.2,random_state=42)
model = RandomForestClassifier()
model.fit(x_train , y_train) 
```

## **Evaluate the model post-correction**

```python
# let's evaluate model

crosss_valid = cross_val_score(model,x_train,y_train,cv=5)
print(f"crosss validation is : {crosss_valid}")
print(f"the score is : {model.score(x_test, y_test)}")
```

```python
# let's find prediction on test data points.
y_pred = model.predict(x_test)
```

```python
# Confusion matrix post-correction
Confusion = confusion_matrix(y_test, y_pred)
Confusion
# wonderfull, the false negative and false positive are now very small.
```
